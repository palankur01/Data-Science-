```python
import pandas as pd

# Build the data
data = {
    'Position': ['Striker', 'Forward', 'Attacking Midfielder', 'Winger', 'Total'],
    'Injured': [45, 56, 24, 20, 145],
    'Not Injured': [32, 38, 11, 9, 90],
    'Total' : [77, 94, 35, 29, 235]
}

#Creating the DataFrame

df1 = pd.DataFrame(data)
df1.set_index('Position', inplace=True)
print(df1)
```

```python
#1.1 what is the probability that a randomly chosen player would suffer an inury? 

# Extract total injured and total players from dataframe
total_injured = df1.loc['Total', 'Injured']
total_players = df1.loc['Total', 'Total']

# Calculate probability
p_injury = total_injured / total_players

# Print result
print(f"Probability that a randomly chosen player would suffer an injury: {p_injury:.4f} ({p_injury*100:.2f}%)")
```

```python
#1.2 What is the probability that a player is a forward or a winger?

# Get total players from the 'Total' row
total_players = df1.loc['Total', 'Total']

# Get Forward total
forward_total = df1.loc['Forward', 'Total']

# Get Winger total
winger_total = df1.loc['Winger', 'Total']

# Calculate probability
p_forward_or_winger = (forward_total + winger_total) / total_players

# Print result
print(f"Probability that a player is Forward or Winger: {p_forward_or_winger:.4f} ({p_forward_or_winger*100:.2f}%)")
```

```python
#1.3 What is the Probability that a randomly chosen player plays in a striker position and has a foot injury? 

# Extract values from the dataframe
striker_injured = df1.loc['Striker', 'Injured']
total_players = df1.loc['Total', 'Total']

# Calculate the probability
p_striker_and_injured = striker_injured / total_players

# Print result
print(f"Probability (Striker and Injured): {p_striker_and_injured:.4f} ({p_striker_and_injured*100:.2f}%)")
```

```python
#1.4 What is the probability that a randomly chosen injured player is a striker?

# Calculate probability
injured_striker = df1.loc['Striker', 'Injured']
total_injured = df1.loc['Total', 'Injured']

p_striker_given_injured = injured_striker / total_injured

print(f"Probability that an injured player is a striker: {p_striker_given_injured:.4f} ({p_striker_given_injured*100:.2f}%)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mu = 5  # mean
sigma = 1.5  # standard deviation

# Create x-axis for plotting
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# Create a plot template
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution')
plt.title('Breaking Strength of Gunny Bags')
plt.xlabel('Breaking Strength (kg/cm)')
plt.ylabel('Density')
```

```python
# 2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per square cm?
p1 = norm.cdf(3.17, mu, sigma)
print(f"2.1 Proportion less than 3.17 kg/cm: {p1:.4f} ({p1*100:.2f}%)")
x_fill = np.linspace(mu - 4*sigma, 3.17, 100)
plt.fill_between(x_fill, norm.pdf(x_fill, mu, sigma), color='skyblue', alpha=0.5, label='Less than 3.17')

plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm?
p2 = 1 - norm.cdf(3.6, mu, sigma)
print(f"2.2 Proportion at least 3.6 kg/cm: {p2:.4f} ({p2*100:.2f}%)")
x_fill = np.linspace(3.6, mu + 4*sigma, 100)
plt.fill_between(x_fill, norm.pdf(x_fill, mu, sigma), color='orange', alpha=0.5, label='At least 3.6')


plt.legend()
plt.grid(True)
plt.show()

```

```python
# 2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 Kg per sq cm?
p3 = norm.cdf(5.5, mu, sigma) - norm.cdf(5, mu, sigma)
print(f"2.3 Proportion between 5 and 5.5 kg/cm: {p3:.4f} ({p3*100:.2f}%)")
x_fill = np.linspace(5, 5.5, 100)
plt.fill_between(x_fill, norm.pdf(x_fill, mu, sigma), color='green', alpha=0.5, label='Between 5 and 5.5')

plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2.4 What proportion of the gunny bags have a breaking stregth NOT between 3 and 7.5 kg per square cm?
p4 = 1 - (norm.cdf(7.5, mu, sigma) - norm.cdf(3, mu, sigma))
print(f"2.4 Proportion NOT between 3 and 7.5 kg/cm: {p4:.4f} ({p4*100:.2f}%)")
x_fill1 = np.linspace(mu - 4*sigma, 3, 100)
x_fill2 = np.linspace(7.5, mu + 4*sigma, 100)
plt.fill_between(x_fill1, norm.pdf(x_fill1, mu, sigma), color='red', alpha=0.5, label='Not between 3 and 7.5')
plt.fill_between(x_fill2, norm.pdf(x_fill2, mu, sigma), color='red', alpha=0.5)

plt.legend()
plt.grid(True)
plt.show()
```

```python
#3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. DO you think Zingaro is justified in thinking so?

import pandas as pd
from scipy.stats import ttest_1samp

# Remove any extra spaces from column names
df.columns = df.columns.str.strip()

# Extract unpolished stone data
unpolished = df['Unpolished'].dropna()

# Perform one-sample t-test
test_stat, p_value = ttest_1samp(unpolished, 150)

# Since we're testing if mean < 150, divide p-value by 2 for one-tailed test
p_value_one_tailed = p_value / 2

print(f"Test Statistic: {test_stat:.4f}")
print(f"One-tailed p-value: {p_value_one_tailed:.4f}")

# Conclusion
alpha = 0.05
if (p_value_one_tailed < alpha) and (test_stat < 0):
    print("Reject H0: Zingaro is justified; unpolished stones may not be suitable.")
else:
    print("Fail to reject H0: Not enough evidence to say unpolished stones are unsuitable.")
```

```python
#4.3 What is the interaction effect between the dentist and method on the hardness of dental implants for each type of alloy?

# Load the Excel file
df = pd.read_csv('/Users/AnkurPal/Downloads/Dental Hardness data.csv')

# Quick check of data
print(df.head())
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# First ensure that columns are read as categorical
df['Dentist'] = df['Dentist'].astype('category')
df['Method'] = df['Method'].astype('category')
df['Alloy'] = df['Alloy'].astype('category')

# First check unique alloys
alloys = df['Alloy'].unique()

for alloy in alloys:
    print(f"\nANALYSIS FOR ALLOY: {alloy}")
    
    # Subset data for this alloy
    subset = df[df['Alloy'] == alloy].copy()
    
    # Make sure subset has categorical columns too
    subset['Dentist'] = subset['Dentist'].astype('category')
    subset['Method'] = subset['Method'].astype('category')
    
    # Perform two-way ANOVA with interaction
    try:
        model = ols('Hardness ~ C(Dentist) * C(Method)', data=subset).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        print(anova_table)
    except Exception as e:
        print(f"Error for Alloy {alloy}: {e}")
```
